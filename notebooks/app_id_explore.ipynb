# Exploratory Data Analysis of app_id table

```python
import pandas as pd 
```

```python
app_id = pd.read_csv("../data/output/app_id.csv")

app_id.info()
```

### Missing data

While it appears there is a lot of missing data in this table, what this table denotes should be a single app_id for each customer so this is expected

We need to find out if any customers are recorded having both android **and** apple, as well as any records with neither


```python
# How many records have no app value 
 
missing_both = (app_id["android_id"].isnull() & app_id["apple_id"].isnull())

print(f"Total rows missing any id: {missing_both.sum()}")

# How many records have 2 ids recorded for a single customer 

missing_neither = (app_id["android_id"].notnull() & app_id["apple_id"].notnull())

print(f"Total rows with 2 recorded ids {missing_neither.sum()}")
```

> Looks good!
