# Exploratory Data Analysis of Address table

```python
import pandas as pd 
```

```python
address = pd.read_csv("../data/output/address.csv")

address.info()
```

### Missing data

Appears to be some missing data in the `district`, `postal_code` and `phone` columns. `address2` is missing completely    


```python
# How many rows are we losing 

total_missing = (address["district"].isnull() | address["postal_code"].isnull() | address["phone"].isnull())
print(f"Total rows missing: {total_missing.sum()}")
```

> For a dataset of 603 rows, 7 missing rows (1.1%) is  certainly a small enough amount to be dropped without a second thought
